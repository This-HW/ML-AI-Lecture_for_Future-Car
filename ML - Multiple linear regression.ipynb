### 식
- Simple linear regression 식은    
H(x) =  W * x + b였다.
- Multiple linear regression 식은    
H(x1, x2, x3) = W1 * x1 + W2 * x2 + w3 * x3 + b 이렇게 된다.
- 항이 여러개 이기에, 행렬 곱, 매트릭스로 표현하는 것이 훨씬 깔끔하다.
>
              (w1)
(x1, x2, x3) * (w2)
              (w3)
 이렇게 표현 할 수있다.<
 
 이제부터 x의 행렬을 X로, w의 행렬을 W로 표현한다.
 이를 간다하게 나타내면,
 H = X * W + b가 된다.

```python
## Multiple linear regression
import tensorflow as tf


## 1. training data set

x_data = [[73, 80, 75],
         [93, 88, 93],
         [89, 91, 90],
         [96, 98, 100],
         [73, 66, 70]]
y_data = [[152], [185], [180], [196], [142]]
# 이제부터는 변수들이 multiple 하기에, 2차원 행렬(matrix)이333다.
# 때문에 모든 데이터들의 형태를 반드시! 2차원 행렬로 맞춰줘야 한다.


## 2. placeholder 생성
X = tf.placeholder(shape=[None, 3], dtype = tf.float32) # 입력값의 열은 3개로 고정이지만, 행은 가변적이다.
Y = tf.placeholder(shape=[None, 1], dtype = tf.float32) # 출력값의 열은 1개로 고정이지만, 행은 가변적이다.


## 3. weight & bias 생성

# W의 초기값을 어떻게 정하는지가 이 알고리즘의 성능에 지대한 영향을 끼친다.
# 하지만 여기에서는 일단 랜덤으로 설정한다.
W = tf.Variable(tf.random_normal([3, 1]), name="weight") # W는 x와 반대로 행 3개(x와 매치), 열 1개다. 
b = tf.Variable(tf.random_normal([1]), name="bias") # b는 1개다. 
# W의 열은 계속 업데이트 되기에 1개면 된다. W와 b는 학습 후 학습된 값 재입력


## 4. Hypothesis
H = tf.matmul(X, W) + b


## 5. cost function (loss function)
cost = tf.reduce_mean(tf.square(H-Y)) # 계속해서 최소제곱법을 사용한다.


## 6. train
# train의 역할은 cost를 계속 줄이는 것이다.
train = tf.train.GradientDescentOptimizer(learning_rate = 0.00001).minimize(cost)
# Linera regression처럼 Gradient Descent사용
# learning rate는 0.001 or 0.01부터 시작해서 cost값을 보고 최적화 해야야 한다.
## -> 3번째 학습 때 0.00001로 수정. -> 그래도 발산한다. -> 인풋값이 잘못되었다.


## 7. session 초기화
sess = tf.Session()
sess.run(tf.global_variables_initializer())


## 8. train 학습
for step in range(3001):
    tmp_train , cost_val = sess.run([train, cost],
                               feed_dict={X:x_data, Y:y_data})
    if step % 300 == 0:
        print("step : ", step, "     cost 값은 : {}".format(cost_val))
```

보통 이런 경우 러닝레이트를 수정해서 재학습을 시킨다.    
러닝레이트를 고쳐서 수정 해 보았지만 결과가 좋지 않았다.    
0.00001까지 수정 했지만, 그래도 발산한다.    

이유?    
-> 이번에는 학습 데이터가 좋지 않았다.    
   이처럼 학습 데이터의 전처리와 정규화는 매우 중요하다.    
   그래서 이번에는 실제 데이터를 이용해 머신러닝을 해 보겠다.
